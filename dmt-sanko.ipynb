```python
import pandas as pd
```

```python
dmtPrice = pd.read_csv('sanko price.csv')
dmtMetrics = pd.read_csv('dmt metrics.csv')
dmtHolders = pd.read_csv('dmtholders.csv')
dmtdau = pd.read_csv('sankodau.csv')
dmtRevenueAndBurn = pd.read_csv('sankorevenueandburn.csv')
dmtTvDau = pd.read_csv('sankotvdau.csv')
```

```python
dmtPrice['Time'] = pd.to_datetime(dmtPrice['Time'], utc=True)
dmtMetrics['Time'] = pd.to_datetime(dmtMetrics['Time'], utc=True)
dmtHolders['Time'] = pd.to_datetime(dmtHolders['Time'], utc=True)
dmtdau['Time'] = pd.to_datetime(dmtdau['Time'], utc=True)
dmtRevenueAndBurn['Time'] = pd.to_datetime(dmtRevenueAndBurn['Time'], utc=True)
dmtTvDau['Time'] = pd.to_datetime(dmtTvDau['Time'], utc=True)
```

```python
dmtPrice['Time'] = dmtPrice['Time'].dt.floor('D')
dmtPrice.groupby('Time')['DMT_price'].mean()
```

```python
df = dmtPrice.merge(dmtMetrics, on='Time', how='inner').merge(dmtHolders, on='Time', how='inner').merge(dmtdau, on='Time', how='inner').merge(dmtRevenueAndBurn, on='Time', how='inner')
```

```python
df_weekly = df.groupby(df['Time'].dt.to_period('W')).mean()
df_weekly.set_index(['Time'], inplace=True)
df_weekly
```

```python
weekly = df_weekly.corr()
```

```python
weekly['DMT_price'].to_csv('weekly.csv')
```

```python
df_monthly = df.groupby(df['Time'].dt.to_period('M')).mean()
df_monthly.set_index(['Time'], inplace=True)
df_monthly
```

```python
monthly = df_monthly.corr()
```

```python
monthly['DMT_price'].to_csv('monthly.csv')
```
